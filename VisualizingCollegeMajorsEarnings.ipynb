```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Make plots displayed inline
%matplotlib inline
```

```python
# Read the dataset into a DataFrame and start exploring the data.
recent_grads = pd.read_csv('recent-grads.csv')
```

```python
# Generate summary statistics for all of the numeric columns.
recent_grads.describe()
```

```python
# Drop rows with missing values. Matplotlib expects that columns of values we pass in have matching lengths and missing values will cause matplotlib to throw errors.
raw_data_count = len(recent_grads.index)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads.index)
```

```python
print(raw_data_count)
print(cleaned_data_count)

# Only one row contained missing values and was dropped.
```

```python
fig1, axes1 = plt.subplots(nrows=3, ncols=2, figsize=(16,16))

recent_grads.plot(ax=axes1[0,0], kind='scatter', x='Sample_size', y='Median', title='Sample_size vs. Median')
recent_grads.plot(ax=axes1[0,1], kind='scatter', x='Sample_size', y='Unemployment_rate', title='Sample_size vs. Unemployment_rate')
recent_grads.plot(ax=axes1[1,0], kind='scatter', x='Full_time', y='Median', title='Full_time vs. Median')
recent_grads.plot(ax=axes1[1,1], kind='scatter', x='ShareWomen', y='Unemployment_rate', title='ShareWomen vs. Unemployment_rate')
recent_grads.plot(ax=axes1[2,0], kind='scatter', x='Men', y='Median', title='Men vs. Median')
recent_grads.plot(ax=axes1[2,1], kind='scatter', x='Women', y='Median', title='Women vs. Median')

plt.tight_layout()
plt.show()
```

```python
fig2, axes2 = plt.subplots(nrows=3, ncols=1, figsize=(8, 12))

recent_grads.plot(ax=axes2[0], kind='scatter', x='Total', y='Median', title='Total vs. Median')
recent_grads.plot(ax=axes2[1], kind='scatter', x='ShareWomen', y='Median', title='ShareWomen vs. Median')
recent_grads.plot(ax=axes2[2], kind='scatter', x='Full_time', y='Median', title='Full_time vs. Median')

plt.tight_layout()
plt.show()
```

* There is a weak positive correlation between the numbero of people in a major and median salary.

* There is a weak negative correlation between share of women and median salary. This suggests that students that majored in subjects that were majority female generally did not make more money.

* There is a weak positive correlation between number of full-time employees and median salary.

```python
fig3, axes3 = plt.subplots(nrows=4, ncols=2, figsize=(16, 12))

ax3a = recent_grads['Sample_size'].hist(ax=axes3[0,0])
ax3b = recent_grads['Median'].hist(ax=axes3[0,1])
ax3c = recent_grads['Employed'].hist(ax=axes3[1,0])
ax3d = recent_grads['Full_time'].hist(ax=axes3[1,1])
ax3e = recent_grads['ShareWomen'].hist(ax=axes3[2,0], bins = 2)
ax3f = recent_grads['Unemployment_rate'].hist(ax=axes3[2,1])
ax3g = recent_grads['Men'].hist(ax=axes3[3,0])
ax3h = recent_grads['Women'].hist(ax=axes3[3,1])

axes3_titles = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
axes3_list = [ax3a, ax3b, ax3c, ax3d, ax3e, ax3f, ax3g, ax3h]

for i, ax in enumerate(axes3_list):
    ax.set_title(axes3_titles[i])
    ax.set_xlabel(axes3_titles[i])
    ax.set_ylabel('Frequency')
    
plt.tight_layout()
plt.show()
```

```python
print('* Count: ' + str(recent_grads['ShareWomen'].count()))
print('* More Women: ' + str(recent_grads['ShareWomen'][recent_grads['ShareWomen'] > .5].count()))
print('* Same Men/Women: ' + str(recent_grads['ShareWomen'][recent_grads['ShareWomen'] == .5].count()))
print('* More Men: ' + str(recent_grads['ShareWomen'][recent_grads['ShareWomen'] < .5].count()))
```

* 44% (76/172) majors are predominantly male, 56% (96/172) are predominantly female.

* The most common median salary range is \$30K - \$40K.
